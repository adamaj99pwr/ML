```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['figure.dpi'] = 140
```

```python
tesla = pd.read_csv('tesla.csv')
tesla.head()
```

```python
tesla.isna().sum()
```

```python
tesla.describe()
```

```python
tesla.info()
```

```python
plt.plot(tesla['Date'], tesla['Close'])
plt.grid()
plt.title('Tesla close price')
tick_inds = np.linspace(0, len(tesla)-1, 8, dtype=int)
plt.xticks(np.array(tesla[tesla.index.isin(tick_inds)]['Date']))
plt.xticks(rotation=45)
plt.show()
```
